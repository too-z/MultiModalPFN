```python
import os
import pandas as pd
```

```python
col_features = [
    'Age',
    'Breed1',
    'Breed2',
    'Gender',
    'Color1',
    'Color2',
    'Color3',
    'MaturitySize',
    'Type',
    'FurLength',
    'Vaccinated',
    'Dewormed',
    'Sterilized',
    'Health',
    'Quantity',
    'Fee',
    'State',
    'VideoAmt',
    'PhotoAmt',
] 
col_exclude = ['PetID', 'RescuerID', 'Description', 'Name']
col_target = 'AdoptionSpeed'
```

```python
train = pd.read_csv('datasets/petfinder-adoption-prediction/train/train.csv')

# create datasets_dir
datasets_dir = 'datasets/petfinder-adoption-prediction'

# For pets that have photos, they will be named in the format of PetID-ImageNumber.jpg. 
# Image 1 is the profile (default) photo set for the pet.
# Always use Image 1

# bring PetID from train
train['PetID'] = train['PetID'].astype(str)
# create a list of all images in the train directory
train_images = [f for f in os.listdir(os.path.join(datasets_dir, 'train_images')) if f.endswith('.jpg')]
# filter images that are in the train set
train_images = [f for f in train_images if f.split('-')[0] in train['PetID'].values]
# create a DataFrame with PetID and ImageNumber
train_images_df = pd.DataFrame({
    'PetID': [f.split('-')[0] for f in train_images],
    'ImageNumber': [f.split('-')[1].split('.')[0] for f in train_images]
})
# keep only ImageNumber 1
train_images_df = train_images_df[train_images_df['ImageNumber'] == '1']
# merge with train DataFrame
train = train.merge(train_images_df, on='PetID', how='left')
# remove rows where ImageNumber is NaN
train = train[train['ImageNumber'].notna()]
# create a new column with the full path to the image
train['ImagePath'] = train['PetID'] + '-1.jpg'
```

```python
x_train = train[col_features]
x_images = train['ImagePath']
y_train = train[col_target]
```
