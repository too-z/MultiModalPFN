```python
import numpy as np
import pandas as pd

# Settings
np.random.seed(42)
num_sets = 6
set_size = (10, 192)
sets = []

# Create the embedding sets
for i in range(num_sets):
    if i < 3:
        # High correlation: similar distributions, little noise
        base = np.random.normal(size=set_size)
        sets.append(base + np.random.normal(scale=0.01, size=set_size))
    else:
        # Low correlation: unrelated random sets
        sets.append(np.random.normal(scale=5.0, size=set_size))

# Function: correlation between sets
def set_correlation(set_a, set_b):
    a, b = set_a.flatten(), set_b.flatten()
    return np.corrcoef(a, b)[0, 1]

# Compute pairwise correlations
corr_matrix = np.zeros((num_sets, num_sets))
for i in range(num_sets):
    for j in range(num_sets):
        print(sets[i].shape, sets[j].shape)
        corr_matrix[i, j] = set_correlation(sets[i], sets[j])

# Visualize or export
corr_df = pd.DataFrame(
    corr_matrix, 
    index=[f'Set_{i}' for i in range(num_sets)], 
    columns=[f'Set_{i}' for i in range(num_sets)]
)
# corr_df.to_csv('corr_matrix.csv')
print(corr_df)
```
